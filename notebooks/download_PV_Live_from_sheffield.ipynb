```python
from datetime import datetime
import pytz
from pvlive_api import PVLive
```

```python
pvl = PVLive()
```

```python
df_from_pv_live_api = pvl.between(
    start=datetime(2022, 6, 1, 0, 30, tzinfo=pytz.utc),
    end=datetime(2023, 4, 26, tzinfo=pytz.utc),
    dataframe=True,
)
```

```python
df_from_pv_live_api.set_index("datetime_gmt").generation_mw.plot()
```
